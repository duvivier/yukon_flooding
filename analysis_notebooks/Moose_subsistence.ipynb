```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import matplotlib.ticker as mticker 
```

```python
# Load your CSV data into a DataFrame called combined_df
combined_df = pd.read_csv('Moose_subsistence.csv')
```

```python
# Clean column names
combined_df.columns = combined_df.columns.str.strip()

# Create plot
plt.figure(figsize=(10, 6))
```

```python
import matplotlib.pyplot as plt

# Clean column names
combined_df.columns = combined_df.columns.str.strip()

# Calculate percent success for each community
combined_df['Fort Yukon Percent Success'] = (
    combined_df['Fort Yukon Successful'] / 
    (combined_df['Fort Yukon Successful'] + combined_df['Fort Yukon Unsuccessful'])
) * 100

combined_df['Tanana Percent Success'] = (
    combined_df['Tanana Successful'] / 
    (combined_df['Tanana Successful'] + combined_df['Tanana Unsuccessful'])
) * 100

combined_df['Eagle Percent Success'] = (
    combined_df['Eagle Successful'] / 
    (combined_df['Eagle Successful'] + combined_df['Eagle Unsuccessful'])
) * 100

combined_df['Galena Percent Success'] = (
    combined_df['Galena Successful'] / 
    (combined_df['Galena Successful'] + combined_df['Galena Unsuccessful'])
) * 100

# Create Plot
plt.plot(combined_df['Year'], combined_df['Fort Yukon Percent Success'],
         color='#ff7f0e', label='Fort Yukon')  # Orange
plt.plot(combined_df['Year'], combined_df['Tanana Percent Success'], 
         color='#2ca02c', label='Tanana')      # Green
plt.plot(combined_df['Year'], combined_df['Eagle Percent Success'], 
         color='#1f77b4', label='Eagle')       # Dark Blue
plt.plot(combined_df['Year'], combined_df['Galena Percent Success'], 
         color='#9467bd', label='Galena')      # Purple

# Add vertical lines for 2009 and 2013
plt.axvline(x=2009, color='red', linestyle='--', linewidth=2)
plt.axvline(x=2013, color='red', linestyle='--', linewidth=2)

plt.gca().xaxis.set_major_locator(mticker.MultipleLocator(5))

# Labels and title
plt.title('Percent Successful Moose Hunts Over Time\n(Fort Yukon, Tanana, Eagle, Galena)')
plt.xlabel('Year')
plt.ylabel('Percent Success (%)')
plt.grid(True)
plt.legend()
plt.tight_layout()
#plt.savefig('Combined_Moose_Success_Total_Counts_Over_Time.pdf', format='pdf')
plt.show()
```
